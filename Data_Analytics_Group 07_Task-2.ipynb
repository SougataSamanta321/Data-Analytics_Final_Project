*Problem Statement:*     Customer Segmentation and Clustering Analysis

*Objective:*     The primary goal of this project is to perform customer segmentation and clustering analysis on a wholesale customer dataset. By leveraging unsupervised learning techniques, specifically the K-means clustering algorithm, we aim to group customers based on their purchasing behavior. Additionally, we explore the characteristics of each cluster to gain insights into distinct customer segments.

*Dataset:*     The dataset contains information about the annual spending on various product categories by different customers. The relevant features for clustering are 'Fresh,' 'Milk,' 'Grocery,' 'Frozen,' 'Detergents_Paper,' and 'Delicassen.' The dataset also includes a 'Region' column, which is not used for clustering.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
df=pd.read_csv("C://Users//91743//Downloads//Wholesale customers data.csv")
```

```python
#Show the dataset of 1st 5 rows
df.head()
```

```python
# Select relevant columns for clustering
columns_for_clustering = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
cluster_data = df[columns_for_clustering]
```

```python
# Standardize the data
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)
```

```python
# Apply k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(cluster_data_scaled)
```

```python
cluster_distribution = df['Cluster'].value_counts().sort_index()
print("Distribution of points in each cluster:")
print(cluster_distribution)
```

```python
# Analyze cluster characteristics
cluster_means = df.groupby('Cluster')[columns_for_clustering].mean()
print("\nCluster characteristics:")
print(cluster_means)
```

```python
raw_data1 = make_blobs(n_samples = 200, n_features = 2)
```

```python
plt.scatter(raw_data1[0][:,0], raw_data1[0][:,1], c=raw_data1[1])
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = [] # within cluster sum of squares

for k in range(1, 15):
    kmeansForLoop = KMeans(n_clusters = k)
    kmeansForLoop.fit(raw_data1[0])
    wcss.append(kmeansForLoop.inertia_)

plt.figure(figsize = (10, 10))
plt.plot(range(1, 15), wcss)
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```
